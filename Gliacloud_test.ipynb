```python
a = []
with open( "raw_sentences.txt", "r",encoding="utf8" ) as f:
    file = f.readlines()
```

```python
sentences = []
tokenizer = RegexpTokenizer(r'\w+')
# tokenizer.tokenize('Eighty-seven miles to go, yet.  Onward!')
for line in file:
    sentence = tokenizer.tokenize(line.lower())
    sentences.append(sentence)
```

```python
sentences
All_words = []
for i in sentences:
    All_words.extend(i)
```

```python
All_words_set = list(set(All_words))
```

```python
window = 2
for i in range(len(All_words)):
    wordpost = []
    wordrear = []
    if i>window:
        if len(All_words)-i > window:
            wordpost = All_words[i-window:i]
            wordrear = All_words[i+1:i+window]
    for j in wordpost:
        d.loc[All_words[i],j]+=1
    for j in wordrear:
        d.loc[All_words[i],j]+=1
        
```
